```python
# Fine-tuning Faster R-CNN on a COCO-style custom dataset

import os
import torch
import torchvision
import cv2
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from pycocotools.coco import COCO
from datetime import datetime
from tqdm import tqdm

# GPU setup
USE_COLAB = False
DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
print("Using device:", DEVICE)

# Paths
BASE_DIR = "/home/manulab/projects/portfolio-work-1"
OUTPUT_DIR = os.path.join(BASE_DIR, "output", datetime.now().strftime("%Y-%m-%d_%H-%M-%S"))
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load class names
with open(os.path.join(BASE_DIR, "class.names")) as f:
    CLASS_NAMES = ["Background"] + [line.strip() for line in f.readlines()]
NUM_CLASSES = len(CLASS_NAMES)

# Helper: transform image

def img_transform(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB).astype(np.float32)
    img /= 255.0
    return torch.from_numpy(img).permute(2, 0, 1)

# Optional augmentations
import torchvision.transforms.functional as F
class Transformation():
    def get_probability(self):
        return np.random.rand() > 0.5

    def random_adjust_contrast(self, image, enable=None):
        enable = self.get_probability() if enable is None else enable
        return F.adjust_contrast(image, 2) if enable else image

    def random_adjust_brightness(self, image, enable=None):
        enable = self.get_probability() if enable is None else enable
        return F.adjust_brightness(image, 2) if enable else image

    def random_hflip(self, image, boxes, enable=None):
        enable = self.get_probability() if enable is None else enable
        if enable:
            new_image = F.hflip(image)
            new_boxes = boxes.clone()
            new_boxes[:, [0, 2]] = image.shape[2] - boxes[:, [2, 0]]
            return new_image, new_boxes
        return image, boxes

# Custom COCO dataset class
class CustomCocoDataset(Dataset):
    def __init__(self, img_dir, ann_path, classes, transforms=None):
        self.img_dir = img_dir
        self.coco = COCO(ann_path)
        self.ids = list(sorted(self.coco.imgs.keys()))
        self.transforms = transforms
        self.classes = classes

    def __getitem__(self, index):
        coco = self.coco
        img_id = self.ids[index]
        ann_ids = coco.getAnnIds(imgIds=img_id)
        anns = coco.loadAnns(ann_ids)
        path = coco.loadImgs(img_id)[0]['file_name']
        img_path = os.path.join(self.img_dir, path)
        img = cv2.imread(img_path)
        img = img_transform(img)

        boxes, labels = [], []
        for ann in anns:
            x, y, w, h = ann['bbox']
            boxes.append([x, y, x+w, y+h])
            labels.append(ann['category_id'])

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)

        if self.transforms:
            img = self.transforms.random_adjust_contrast(img)
            img = self.transforms.random_adjust_brightness(img)
            img, boxes = self.transforms.random_hflip(img, boxes)

        target = {"boxes": boxes, "labels": labels, "image_id": torch.tensor([img_id])}
        return img, target

    def __len__(self):
        return len(self.ids)

# Load datasets
train_dataset = CustomCocoDataset(
    os.path.join(BASE_DIR, "train"),
    os.path.join(BASE_DIR, "train/_annotations.coco.json"),
    CLASS_NAMES,
    transforms=Transformation()
)
valid_dataset = CustomCocoDataset(
    os.path.join(BASE_DIR, "valid"),
    os.path.join(BASE_DIR, "valid/_annotations.coco.json"),
    CLASS_NAMES
)
test_dataset = CustomCocoDataset(
    os.path.join(BASE_DIR, "test"),
    os.path.join(BASE_DIR, "test/_annotations.coco.json"),
    CLASS_NAMES
)

def collate_fn(batch):
    return tuple(zip(*batch))

data_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)
data_loader_test = DataLoader(valid_dataset, batch_size=2, shuffle=False, collate_fn=collate_fn)

# Load pre-trained model and adapt classifier
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights='COCO_V1')
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, NUM_CLASSES)
model.to(DEVICE)

# Training setup
optimizer = torch.optim.SGD([p for p in model.parameters() if p.requires_grad], lr=0.005, momentum=0.9, weight_decay=0.0005)
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)

# Training and evaluation loops
def train_one_epoch(model, optimizer, data_loader, device):
    model.train()
    losses = []
    for images, targets in tqdm(data_loader):
        images = list(img.to(device) for img in images)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
        loss_dict = model(images, targets)
        loss = sum(loss for loss in loss_dict.values())
        losses.append(loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    return losses

def evaluate(model, data_loader, device):
    model.train()  # <- Set to train mode to compute losses!
    losses = []
    for images, targets in tqdm(data_loader):
        images = list(img.to(device) for img in images)
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]
        with torch.no_grad():
            loss_dict = model(images, targets)
            loss = sum(loss for loss in loss_dict.values())
            losses.append(loss.item())
    return losses


# Run training
train_losses, valid_losses = [], []
for epoch in range(1000):
    print(f"Epoch {epoch+1} / 5")
    train_loss = train_one_epoch(model, optimizer, data_loader, DEVICE)
    val_loss = evaluate(model, data_loader_test, DEVICE)
    lr_scheduler.step()
    train_losses.extend(train_loss)
    valid_losses.extend(val_loss)
    ckpt_path = os.path.join(OUTPUT_DIR, f"epoch_{epoch+1}.pth")
    torch.save(model.state_dict(), ckpt_path)

# Plot losses
plt.figure()
plt.plot(train_losses, label="Train")
plt.plot(valid_losses, label="Valid")
plt.legend()
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.savefig(os.path.join(OUTPUT_DIR, "loss_plot.png"))
plt.show()
```

```python
# ✅ Clean Inference and Evaluation Code (Paste into Notebook)

import os
import torch
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import random
from tqdm import tqdm
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import json
import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor

# Define once — points to your trained models
OUTPUT_DIR = "/home/manulab/projects/portfolio-work-1/output/2025-04-21_11-39-06"
BASE_DIR = "/home/manulab/projects/portfolio-work-1"

# Load your classes
with open(os.path.join(BASE_DIR, "class.names")) as f:
    CLASS_NAMES = ["Background"] + [line.strip() for line in f.readlines()]

# Helper: draw boxes

def plot_image(img, boxes, scores, labels, dataset, save_path=None, threshold=0.5):
    cmap = plt.get_cmap("tab20b")
    class_labels = np.array(dataset)
    colors = [cmap(i) for i in np.linspace(0, 1, len(class_labels))]

    img = img.permute(1, 2, 0).cpu().numpy()
    img = np.clip(img, 0, 1)

    fig, ax = plt.subplots(1, figsize=(16, 8))
    ax.imshow(img)

    for i, box in enumerate(boxes):
        if scores[i] < threshold:
            continue

        class_pred = labels[i]
        conf = scores[i]
        width = box[2] - box[0]
        height = box[3] - box[1]

        rect = patches.Rectangle(
            (box[0], box[1]), width, height,
            linewidth=2,
            edgecolor=colors[int(class_pred)],
            facecolor="none"
        )
        ax.add_patch(rect)
        ax.text(
            box[0], box[1] - 5,
            f"{class_labels[int(class_pred)]}: {conf:.2f}",
            color="white",
            backgroundcolor=colors[int(class_pred)]
        )

    if save_path:
        plt.savefig(save_path)

    plt.show()

# Define the device (CPU or GPU)
DEVICE = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# Load trained model checkpoint
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights='COCO_V1')
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = FastRCNNPredictor(in_features, len(CLASS_NAMES))
model.load_state_dict(torch.load(os.path.join(OUTPUT_DIR, sorted([f for f in os.listdir(OUTPUT_DIR) if f.endswith(".pth")])[-1])))
model.to(DEVICE)
model.eval()

# Load the test dataset
from torchvision.datasets import CocoDetection
from torchvision.transforms import ToTensor

test_dataset = CocoDetection(
    root=os.path.join(BASE_DIR, "test"),
    annFile=os.path.join(BASE_DIR, "test/_annotations.coco.json"),
    transform=ToTensor()
)

# Run inference and save images
save_dir = os.path.join(OUTPUT_DIR, "inference_results")
os.makedirs(save_dir, exist_ok=True)

N = 5
threshold = 0.5
for i in range(N):
    idx = random.randint(0, len(test_dataset) - 1)
    img, _ = test_dataset[idx]
    img_input = img.to(DEVICE).unsqueeze(0)

    with torch.no_grad():
        pred = model(img_input)[0]

    boxes = pred['boxes'].cpu()
    scores = pred['scores'].cpu()
    labels = pred['labels'].cpu()

    filename = f"inference_{i:04d}.png"
    save_path = os.path.join(save_dir, filename)
    plot_image(img, boxes, scores, labels, CLASS_NAMES, save_path=save_path, threshold=threshold)

print(f"Saved {N} inference images to: {save_dir}")

# --------------------------------------------
# 🧠 Compute mAP, Precision, and Recall (COCO API)
# --------------------------------------------

def convert_to_coco_format(model, dataset, device):
    model.eval()
    results = []
    print("Generating predictions for evaluation...")
    for idx in tqdm(range(len(dataset))):
        img, _ = dataset[idx]
        img_input = img.to(device).unsqueeze(0)
        with torch.no_grad():
            outputs = model(img_input)[0]

        boxes = outputs['boxes'].cpu().numpy()
        scores = outputs['scores'].cpu().numpy()
        labels = outputs['labels'].cpu().numpy()
        image_id = int(dataset.coco.imgs[dataset.ids[idx]]['id'])

        for box, score, label in zip(boxes, scores, labels):
            x_min, y_min, x_max, y_max = box
            width = x_max - x_min
            height = y_max - y_min

            result = {
                "image_id": image_id,
                "category_id": int(label),
                "bbox": [float(x_min), float(y_min), float(width), float(height)],
                "score": float(score)
            }
            results.append(result)
    return results

def evaluate_coco_predictions(predictions, ann_file, output_dir):
    pred_path = os.path.join(output_dir, "predictions.json")
    with open(pred_path, "w") as f:
        json.dump(predictions, f)

    coco_gt = COCO(ann_file)
    coco_dt = coco_gt.loadRes(pred_path)
    coco_eval = COCOeval(coco_gt, coco_dt, iouType='bbox')
    coco_eval.evaluate()
    coco_eval.accumulate()
    coco_eval.summarize()
    return coco_eval

print("\n--- Running mAP/Precision/Recall Evaluation on Test Set ---")
predictions = convert_to_coco_format(model, test_dataset, DEVICE)
evaluate_coco_predictions(
    predictions,
    ann_file=os.path.join(BASE_DIR, "test/_annotations.coco.json"),
    output_dir=OUTPUT_DIR
)
```
